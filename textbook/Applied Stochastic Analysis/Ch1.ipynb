```python
import numpy as np
import matplotlib.pyplot as plt


# we set n= 10,100, 1000 as an example, variance as 4^2
A = np.zeros([10,10])
for i in range(10):
    for j in range(10):
        A[i,j] = np.random.normal(0,4)
B = A / np.sqrt(10*4)

a,b = np.linalg.eig(B)

x = np.real(a)
y = np.imag(a)

plt.scatter(x,y)
plt.axis('equal')
plt.show()

```
